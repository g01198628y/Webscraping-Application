```python
import requests
from bs4 import BeautifulSoup

r = requests.get("http://pythonhow.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/")
c = r.content

soup = BeautifulSoup(c,"html.parser")

all = soup.find_all("div",{"class":"propertyRow"})

all[0].find("h4",{"class":"propPrice"}).text.encode('utf-8').replace("\n","").replace(" ","")
```

```python
l = []
for item in all:
    d = {}
    
    d["Address"]=item.find_all("span",{"class","propAddressCollapse"})[0].text.encode('utf-8')
    d["Locality"]=item.find_all("span",{"class","propAddressCollapse"})[1].text.encode('utf-8')
    d["Price"]=item.find("h4",{"class":"propPrice"}).text.encode('utf-8').replace("\n","").replace(" ","")
    try:
        d["Beds"]=item.find("span",{"class","infoBed"}).find("b").text.encode('utf-8')
    except:
        d["Beds"]=None
        
    try:
        d["Area"]=item.find("span",{"class","infoSqft"}).find("b").text.encode('utf-8')
    except:
        d["Area"]=None 
        
    try:
        d["Full Baths"]=item.find("span",{"class","infoValueFullBath"}).find("b").text.encode('utf-8')
    except:
        d["Full Baths"]=None 
        
    try:
        d["Half Baths"]=item.find("span",{"class","infoValueHalfBath"}).find("b").text.encode('utf-8')
    except:
        d["Half Baths"]=None 
        
    for column_group in item.find_all("div",{"class":"columnGroup"}):
        for feature_group, feature_name in zip(column_group.find_all("span",{"class":"featureGroup"}),column_group.find_all("span",{"class":"featureName"})):
            if "Lot Size" in feature_group.text:
                d["Lot Size"]=feature_name.text.encode('utf-8')
    l.append(d)
```

```python
l
```

```python
import pandas
df = pandas.DataFrame(l)
```
